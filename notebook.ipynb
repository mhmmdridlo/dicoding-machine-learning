# Data Cleaning & Transformation

## Data Train

```python
# cek info data
train.info()
```

```python
# cek statistik deskriptif
train.describe(include='all')
```

```python
# cek missing value
missing_values = train.isnull().sum()
missing_values[missing_values > 0]
```

Karena jumlah data hanya sekitar 1460 baris dan missing value yang didapat rata-rata lebih dari 1000, maka tidak mungkin kita untuk langsung menghapusnya

Maka dari itu, kita akan mencoba untuk memisahkan kolom yang memiliki missing value lebih dari 75% dan kurang dari 75%

```python
# memisahkan missing value
less = missing_values[missing_values < 1000].index
over = missing_values[missing_values > 1000].index
print('missing value berhasi dipisahkan')
```

Berdasarkan dari hasil tersebut, kita akan menggunakan beberapa langkah untuk menangani nilai yang hilang

### Mengisi Nilai yang Hilang

Karena beberapa kolom memiliki missing value dibawah 75%, jadi kita akan mengisi nilainya

Akan tetapi, karena kolom tersebut memiliki tipe data yang berbeda, jadi saat mengisinya akan kita bagi ke dua bagian

#### Kolom Tipe Data Numerik

```python
# mengisi nilai yang hilang dengan median untuk kolom numerik
numeric_features = train[less].select_dtypes(include=['number']).columns
train[numeric_features] = train[numeric_features].fillna(train[numeric_features].median())

print('berhasil mengisi nilai untuk kolom numerik')
```

#### Kolom Tipe Data String/Object

Selanjutnya kita akan mengisi missing value pada kolom yang memiliki tipe data string/object

Sedikit berbeda dengan kolom tipe data numerik, pada tipe data string/object kita tidak dapat menggunakan fungsi aggregasi (ex mean, median)

Biasanya ada dua cara yang sering dilakukan untuk mengatasi ini
- Mengisi dengan modus (nilai yang sering muncul)
- Mengisi dengan kategori baru (missing atau unknown)

Untuk kasus ini, kita akan mengatasinya dengan modus atau mengisinya dengan nilai yang sering muncul

```python
# mengisi nilai yang hilang dengan modus untuk kolom kategori
kategorical_features = train[less].select_dtypes(include=['object']).columns

for column in kategorical_features:
    train[column] = train[column].fillna(train[column].mode()[0])

print('berhasil mengisi nilai untuk kolom kategorical')
```

### Menghapus Nilai yang Hilang

Karena kita tadi sudah menerapkan batasan 75% dan data yang dibawahnya sudah kita isi, maka untuk data yang melebihi akan kita hapus

```python
# menghapus kolom dengan missing value terbanyak
df = train.drop(columns=over)

print('data berhasil dihapus')
```

```python
# memeriksa hasil penanganan missing value
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

### Mengatasi Outlier

```python
for feature in numeric_features:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[feature])
    plt.title(f'Box plot of {feature}')
    plt.show()
```

Dari diagram boxplot yang ditampilkan, nilai yang berada di bawah batas bawah atau di atas batas atas dianggap sebagai sebauh outlier.

Untuk mengatasi permasalahan ini, ada dua pilihan
- Menghapus outlier
- Menggantinya dengan nilai moderat (atau batas terdekat)

Pada kasus ini, kita akan menanganinya dengan metode IQR (Interquartile Range)

```python
# mengidentifikasi outlier dengan menggunakan IQR
Q1 = df[numeric_features].quantile(0.25)
Q3 = df[numeric_features].quantile(0.75)
IQR = Q3-Q1
```
